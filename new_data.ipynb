```python
import pandas as pd

# Fish types and their features
fish_types = [
    "Gilt Head Bream", "Red Sea Bream", "Sea Bass", "Red Mullet", "Horse Mackerel",
    "Black Sea Sprat", "Striped Red Mullet", "Trout", "Shrimp"
]

# Features of fishes
features = ["Size", "Shape", "Colour", "Weight", "Height", "Lifespan", "Production", "Live Stock", "Purchases", "Seasonal Data"]

# Generating dummy data for illustration
import numpy as np
np.random.seed(0)  # For reproducibility

# Generating data, cycling through the fish_types to ensure all are included without skipping
data = []
for i in range(60):
    fish_data = [
        fish_types[i % len(fish_types)],  # Name
        np.random.choice(["Small", "Medium", "Large"]),  # Size
        np.random.choice(["Elongated", "Round", "Oval"]),  # Shape
        np.random.choice(["Silver", "Red", "Blue", "Green"]),  # Colour
        f"{np.random.uniform(0.5, 5.0):.2f} kg",  # Weight
        f"{np.random.uniform(10.0, 100.0):.2f} cm",  # Height
        f"{np.random.randint(1, 10)} years",  # Lifespan
        f"{np.random.randint(100, 1000)} tons",  # Production
        f"{np.random.randint(10, 100)}K",  # Live Stock
        f"{np.random.randint(1, 500)} units",  # Purchases
        np.random.choice(["Spring", "Summer", "Autumn", "Winter"])  # Seasonal Data
    ]
    data.append(fish_data)

# Creating DataFrame
df_fish_data = pd.DataFrame(data, columns=["Fish Type"] + features)

df_fish_data.head(60)
```

```python
df_fish_data.to_csv("fish_analysis.csv")
```

```python
df_fish_data=pd.read_csv("fish_analysis.csv")
```

```python
import matplotlib.pyplot as plt

def plot_fish_demand(fish_name, df):
    """
    Plot the demand (Purchases) of the specified fish type over the dataset.

    Parameters:
    - fish_name: Name of the fish to plot demand for.
    - df: DataFrame containing the fish data.
    """
    # Filter the DataFrame for the specified fish type
    fish_data = df[df["Fish Type"] == fish_name]
    
    # Extracting purchases and converting to numerical value
    purchases = fish_data["Purchases"].apply(lambda x: int(x.split()[0]))
    
    # Plot
    plt.figure(figsize=(10, 6))
    plt.plot(purchases.reset_index(drop=True), marker='o', linestyle='-', color='blue')
    plt.title(f'Demand for {fish_name}', fontsize=15)
    plt.xlabel('Data Point', fontsize=12)
    plt.ylabel('Purchases (Units)', fontsize=12)
    plt.grid(True)
    plt.show()

# Example: Plotting the demand for Gilt Head Bream
plot_fish_demand("Horse Mackerel", df_fish_data)
```

```python
# Let's choose another fish type for a more detailed demand analysis.
# This time, we'll analyze the demand for "Trout" and include seasonal data in the analysis.

def plot_fish_demand_detailed(fish_name, df):
    """
    Plot the demand (Purchases) of the specified fish type over the dataset, including seasonal data.

    Parameters:
    - fish_name: Name of the fish to plot demand for.
    - df: DataFrame containing the fish data.
    """
    # Filter the DataFrame for the specified fish type
    fish_data = df[df["Fish Type"] == fish_name]
    
    # Extracting purchases and converting to numerical value
    purchases = fish_data["Purchases"].apply(lambda x: int(x.split()[0]))
    seasons = fish_data["Seasonal Data"]
    
    # Plot
    plt.figure(figsize=(12, 8))
    for season in seasons.unique():
        season_data = fish_data[fish_data["Seasonal Data"] == season]
        season_purchases = season_data["Purchases"].apply(lambda x: int(x.split()[0]))
        plt.plot(season_purchases.reset_index(drop=True), marker='o', linestyle='-', label=season)
        
    plt.title(f'Detailed Demand Analysis for {fish_name}', fontsize=15)
    plt.xlabel('Data Point', fontsize=12)
    plt.ylabel('Purchases (Units)', fontsize=12)
    plt.legend(title="Season")
    plt.grid(True)
    plt.show()

# Plotting the detailed demand for Trout
plot_fish_demand_detailed("Horse Mackerel", df_fish_data)
```
